```python
#Dataframe of aggregated Transactions
path =r"C:\Users\saran\Downloads\pulse-master\pulse-master\data\aggregated\transaction\country\india\state"
agg_trans_list = os.listdir(path)

columns = {'State': [], 'Year': [], 'Quarter': [], 'Transaction_type': [], 'Transaction_count': [],
            'Transaction_amount': []}
for state in agg_trans_list:
    cur_state = path + "/"+ state
    agg_year_list = os.listdir(cur_state)
    
    for year in agg_year_list:
        cur_year = cur_state + "/"+ year 
        agg_file_list = os.listdir(cur_year)
        
        for file in agg_file_list:
            cur_file = cur_year +"/"+ file
            data = open(cur_file, 'r')
            A = json.load(data)
            
            for i in A['data']['transactionData']:
                name = i['name']
                count = i['paymentInstruments'][0]['count']
                amount = i['paymentInstruments'][0]['amount']
                columns['Transaction_type'].append(name)
                columns['Transaction_count'].append(count)
                columns['Transaction_amount'].append(amount)
                columns['State'].append(state)
                columns['Year'].append(year)
                columns['Quarter'].append(int(file.strip('.json')))
                
df_agg_trans = pd.DataFrame(columns)
```

```python
#Dataframe of aggregated user
path =r"C:/Users/saran/Downloads/pulse-master/pulse-master/data/aggregated/user/country/india/state" 

agg_user_list = os.listdir(path)

columns1 = {'State': [], 'Year': [], 'Quarter': [], 'Brands': [], 'Count': [],
            'Percentage': []}
for state in agg_user_list:
    cur_state = path +"/" + state 
    agg_year_list = os.listdir(cur_state)
    
    for year in agg_year_list:
        cur_year = cur_state+"/"  + year +"/"
        agg_file_list = os.listdir(cur_year)

        for file in agg_file_list:
            cur_file = cur_year +"/" + file
            data = open(cur_file, 'r')
            B = json.load(data)
            try:
                for i in B["data"]["usersByDevice"]:
                    
                    brand_name = i["brand"]
                    counts = i["count"]
                    percents = i["percentage"]
                    columns1["Brands"].append(brand_name)
                    columns1["Count"].append(counts)
                    columns1["Percentage"].append(percents)
                    columns1["State"].append(state)
                    columns1["Year"].append(year)
                    columns1["Quarter"].append(int(file.strip('.json')))
            except:
                pass
df_agg_user = pd.DataFrame(columns1)
```

```python
#Dataframe of map transactions
path2 =r"C:/Users/saran/Downloads/pulse-master/pulse-master/data/map/transaction/hover/country/india/state" 
 
map_trans_list = os.listdir(path2)

columns2 = {'State': [], 'Year': [], 'Quarter': [], 'District': [], 'Count': [],
            'Amount': []}

for state in map_trans_list:
    cur_state = path2+"/" + state+"/" 
    map_year_list = os.listdir(cur_state)
    
    for year in map_year_list:
        cur_year = cur_state+ "/" + year 
        map_file_list = os.listdir(cur_year)
        
        for file in map_file_list:
            cur_file = cur_year+ "/" + file
            data = open(cur_file,'r')
            C = json.load(data)
            
            for i in C["data"]["hoverDataList"]:
            
                district = i["name"]
                count = i["metric"][0]["count"]
                amount = i["metric"][0]["amount"]
                columns2["District"].append(district)
                columns2["Count"].append(count)
                columns2["Amount"].append(amount)
                columns2['State'].append(state)
                columns2['Year'].append(year)
                columns2['Quarter'].append(int(file.strip('.json')))
                
df_map_trans = pd.DataFrame(columns2)
```

```python
#Dataframe of map user

path3 =r"C:/Users/saran/Downloads/pulse-master/pulse-master/data/map/user/hover/country/india/state"  

map_user_list = os.listdir(path3)

columns3 = {"State": [], "Year": [], "Quarter": [], "District": [],
            "RegisteredUser": [], "AppOpens": []}

for state in map_user_list:
    cur_state = path3+ "/" + state 
    map_year_list = os.listdir(cur_state)
    
    for year in map_year_list:
        cur_year = cur_state+ "/" + year 
        map_file_list = os.listdir(cur_year)
        
        for file in map_file_list:
            cur_file = cur_year+ "/" + file
            data = open(cur_file, 'r')
            D = json.load(data)
            
            for D in D["data"]["hoverData"].items():
                District = D[0]
                registeredUser = D[1]["registeredUsers"]
                appOpens = D[1]['appOpens']
                columns3["District"].append(District)
                columns3["RegisteredUser"].append(registeredUser)
                columns3["AppOpens"].append(appOpens)
                columns3['State'].append(state)
                columns3['Year'].append(year)
                columns3['Quarter'].append(int(file.strip('.json')))
                
df_map_user = pd.DataFrame(columns3)
```

```python
#top transactions
path4 = r"C:/Users/saran/Downloads/pulse-master/pulse-master/data/top/transaction/country/india/state"

top_trans_list = os.listdir(path4)
columns4 = {'State': [], 'Year': [], 'Quarter': [], 'Pincode': [], 'Transaction_count': [],
            'Transaction_amount': []}

for state in top_trans_list:
    cur_state = path4 + "/" + state 
    top_year_list = os.listdir(cur_state)
    
    for year in top_year_list:
        cur_year = cur_state + "/"+ year 
        top_file_list = os.listdir(cur_year)
        
        for file in top_file_list:
            cur_file = cur_year + "/"+ file
            data = open(cur_file, 'r')
            E = json.load(data)
            
            for i in E['data']['pincodes']:
                name = i['entityName']
                count = i['metric']['count']
                amount = i['metric']['amount']
                columns4['Pincode'].append(name)
                columns4['Transaction_count'].append(count)
                columns4['Transaction_amount'].append(amount)
                columns4['State'].append(state)
                columns4['Year'].append(year)
                columns4['Quarter'].append(int(file.strip('.json')))
df_top_trans = pd.DataFrame(columns4)
```

```python
def insights_page(select):
    st.title(':violet[BASIC INSIGHTS]')
    st.subheader("The basic insights are derived from the Analysis of the Phonepe Pulse data. It provides a clear idea about the analysed data.")
    options = ["--select--",
            "1. Top 10 states based on year and amount of transaction",
            "2. Least 10 states based on year and amount of transaction",
            "3. Top 10 States and Districts based on Registered_users",
            "4. Least 10 States and Districts based on Registered_users",
            "5. Top 10 Districts based on the Transaction Amount",
            "6. Least 10 Districts based on the Transaction Amount",
            "7. Top 10 Districts based on the Transaction count",
            "8. Least 10 Districts based on the Transaction count",
            "9. Top Transaction types based on the Transaction Amount",
            "10. Top 10 Mobile Brands based on the User count of transaction"]
    select = st.selectbox(":violet[Select the option]",options)
```

```python
if 'select' == "1. Top 10 states based on year and amount of transaction":
        cursor.execute(
            "SELECT DISTINCT State,Year, SUM(Transaction_amount) AS Total_Transaction_Amount FROM top_trans GROUP BY State,Year ORDER BY Total_Transaction_Amount DESC LIMIT 10");

        data = cursor.fetchall()
        columns = ['States', 'Year', 'Transaction_amount']
        df = pd.DataFrame(data, columns=columns, index=range(1, len(data) + 1))

        col1, col2 = st.columns(2)
        with col1:
            st.write(df)
        with col2:
            st.title("Top 10 states based on amount of transaction")
            st.bar_chart(data=df, x="Transaction_amount", y="States")
```

```python
 if 'select' == "2. Least 10 states based on year and amount of transaction":
        cursor.execute(
            "SELECT DISTINCT State,Year, SUM(Transaction_amount) as Total FROM top_trans GROUP BY State, Year ORDER BY Total ASC LIMIT 10");
        data = cursor.fetchall()
        columns = ['States', 'Year', 'Transaction_amount']
        df = pd.DataFrame(data, columns=columns, index=range(1,len(data)+1))
        col1, col2 = st.columns(2)
        with col1:
            st.write(df)
        with col2:
            st.title("Least 10 states based on amount of transaction")
            st.bar_chart(data=df, x="Transaction_amount", y="States")
```
